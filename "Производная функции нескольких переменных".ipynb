Задание. Вычислить градиент функции в точке (1,2)

```python
#f(x,y) = 2*x**2*y**3+1/x+y**2*x+7 дана такая функция
```

```python
import numdifftools as nd
```

```python
def rosen(x): #Определила функцию через def
    return 2*x[0]**2*x[1]**3+1/x[0]+x[1]**2*x[0]+7

grad2 = nd.Gradient(rosen)([1, 2]) #Нашла градиент функции с помощью Gradient в точке (1, 2)
grad2
```

Задание. Построить трехмерный график функции

```python
#f(x) = y**2 - x**2*y + 2*x*y дана такая функция
```

```python
from mpl_toolkits import mplot3d #Установила нужные библиотеки
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
```

```python
%matplotlib inline
```

```python
def f(x, y): #Определила функцию через def
    return y**2 - x**2*y + 2*x*y

x = np.linspace(-6, 6, 30) #Область определения функции
y = np.linspace(-6, 6, 30)#Область значений функции

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d') #Построила трехмерный график заданной функции
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
